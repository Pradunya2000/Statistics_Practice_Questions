### 1. Generate a list of 100 integers containing values between 90 to 130 and store it in the variable `int_list`.
After generating the list, find the following:
(i) Write a Python function to calculate the mean of a given list of numbers.
Create a function to find the median of a list of numbers.
(ii) Develop a program to compute the mode of a list of integers.
(iii) Implement a function to calculate the weighted mean of a list of values and their corresponding weights.
(iv) Write a Python function to find the geometric mean of a list of positive numbers.
(v) Create a program to calculate the harmonic mean of a list of values.
(vi) Build a function to determine the midrange of a list of numbers (average of the minimum and maximum).
(vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of
outliers.

```python
import random

# Generating a list of 100 integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# (i) Functions to Calculate Mean and Median
# mean

def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

# median
def calculate_median(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    mid = n // 2

    if n % 2 == 0:
        return (sorted_numbers[mid - 1] + sorted_numbers[mid]) / 2
    else:
        return sorted_numbers[mid]

# (ii) Program to Compute the Mode

from collections import Counter

def calculate_mode(numbers):
    frequency = Counter(numbers)
    mode_data = frequency.most_common()
    max_frequency = mode_data[0][1]

    modes = [num for num, freq in mode_data if freq == max_frequency]
    return modes

# (iii) Function to Calculate Weighted Mean

def calculate_weighted_mean(values, weights):
    weighted_sum = sum(value * weight for value, weight in zip(values, weights))
    total_weights = sum(weights)
    return weighted_sum / total_weights

# (iv) Function to Find the Geometric Mean

import math

def calculate_geometric_mean(numbers):
    
    product = math.prod(numbers)
    return product ** (1 / len(numbers))
    
# (v) Function to Calculate the Harmonic Mean

def calculate_harmonic_mean(numbers):
    n = len(numbers)
    return n / sum(1 / number for number in numbers)

# (vi) Function to Determine the Midrange

def calculate_midrange(numbers):
    return (min(numbers) + max(numbers)) / 2

# (vii) Program to Find the Trimmed Mean

def calculate_trimmed_mean(numbers, percentage=10):
    sorted_numbers = sorted(numbers)
    trim_count = int(len(numbers) * percentage / 100)

    trimmed_numbers = sorted_numbers[trim_count: -trim_count] if trim_count > 0 else sorted_numbers
    return calculate_mean(trimmed_numbers)

# Assuming 'int_list' is generated above
print("Mean:", calculate_mean(int_list))
print("Median:", calculate_median(int_list))
print("Mode:", calculate_mode(int_list))
print("Weighted Mean:", calculate_weighted_mean(int_list, [1] * len(int_list)))  # Using equal weights as an example
print("Geometric Mean:", calculate_geometric_mean([num for num in int_list if num > 0]))  # Geometric mean needs positive numbers
print("Harmonic Mean:", calculate_harmonic_mean(int_list))
print("Midrange:", calculate_midrange(int_list))
print("Trimmed Mean (10%):", calculate_trimmed_mean(int_list, 10))
```

### 3. Write a Python class representing a discrete random variable with methods to calculate its expected value and variance.

```python
class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        
        if len(values) != len(probabilities):
            raise ValueError("Values and probabilities lists must be of the same length.")
        if not (0 <= sum(probabilities) <= 1):
            raise ValueError("Probabilities must sum to 1.")
        
        self.values = values
        self.probabilities = probabilities

    def expected_value(self):
    
        return sum(value * prob for value, prob in zip(self.values, self.probabilities))

    def variance(self):
       
        expected_val = self.expected_value()
        return sum(prob * (value - expected_val) ** 2 for value, prob in zip(self.values, self.probabilities))
```

### 5. Create a Python function to generate random samples from a given probability distribution (e.g., binomial, Poisson) and calculate their mean and variance.

```python
import numpy as np
from scipy.stats import binom, poisson

def generate_samples(distribution, params, sample_size=1000):
    
    if distribution == 'binomial':
        n, p = params['n'], params['p']
        samples = binom.rvs(n, p, size=sample_size)
    elif distribution == 'poisson':
        lambda_ = params['lambda_']
        samples = poisson.rvs(lambda_, size=sample_size)
    else:
        raise ValueError("Unsupported distribution. Choose 'binomial' or 'poisson'.")
    
    # Calculate mean and variance of the samples
    mean = np.mean(samples)
    variance = np.var(samples)
    
    return mean, variance

# Example usage:
# Generate samples from a binomial distribution
binomial_params = {'n': 10, 'p': 0.5}
binomial_mean, binomial_variance = generate_samples('binomial', binomial_params)
print("Binomial Distribution - Mean:", binomial_mean, ", Variance:", binomial_variance)

# Generate samples from a Poisson distribution
poisson_params = {'lambda_': 4}
poisson_mean, poisson_variance = generate_samples('poisson', poisson_params)
print("Poisson Distribution - Mean:", poisson_mean, ", Variance:", poisson_variance)
```

### 6.Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute the mean, variance, and standard deviation of the samples.

```python
import numpy as np

def gaussian_statistics(mean=0, std_dev=1, sample_size=1000):
 
    # Generate random samples from a Gaussian distribution
    samples = np.random.normal(loc=mean, scale=std_dev, size=sample_size)
    
    # Calculate statistics
    sample_mean = np.mean(samples)
    sample_variance = np.var(samples)
    sample_std_dev = np.std(samples)
    
    return sample_mean, sample_variance, sample_std_dev

# Example usage:
mean, variance, std_dev = gaussian_statistics(mean=0, std_dev=1, sample_size=1000)
print("Sample Mean:", mean)
print("Sample Variance:", variance)
print("Sample Standard Deviation:", std_dev)
```

### 8. Write a Python function to calculate the probability density function (PDF) of a continuous random variable for a given normal distribution.

```python
import numpy as np
from scipy.stats import norm

def normal_pdf(x, mean=0, std_dev=1):
  
    # Using the formula for the normal distribution's PDF
    pdf_value = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-((x - mean) ** 2) / (2 * std_dev ** 2))
    
    # Alternatively, using scipy.stats.norm for verification
    pdf_value_scipy = norm.pdf(x, loc=mean, scale=std_dev)
    
    return pdf_value, pdf_value_scipy  # Returning both values for comparison

# Example usage:
x_value = 1.5
mean = 0
std_dev = 1
pdf_result, pdf_result_scipy = normal_pdf(x_value, mean, std_dev)
print("PDF calculated manually:", pdf_result)
print("PDF using scipy:", pdf_result_scipy)
```

### 9. Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.

```python
import numpy as np

def exponential_cdf(x, rate):

    if x < 0:
        return 0  # CDF is 0 for negative x
    if rate <= 0:
        raise ValueError("Rate parameter must be positive.")
    
    cdf_value = 1 - np.exp(-rate * x)
    return cdf_value

# Example usage:
x_value = 2.0
rate_parameter = 0.5
cdf_result = exponential_cdf(x_value, rate_parameter)
print(f"CDF of exponential distribution at x = {x_value} (rate = {rate_parameter}):", cdf_result)
```

### 10. Write a Python function to calculate the probability mass function (PMF) of Poisson distribution.

```python
import numpy as np
import math

def poisson_pmf(k, rate):

    if k < 0 or not isinstance(k, int):
        raise ValueError("k must be a non-negative integer.")
    if rate < 0:
        raise ValueError("Rate parameter must be non-negative.")
    
    pmf_value = (np.exp(-rate) * (rate ** k)) / math.factorial(k)
    return pmf_value

# Example usage:
k_value = 3
rate_parameter = 2.5
pmf_result = poisson_pmf(k_value, rate_parameter)
print(f"PMF of Poisson distribution at k = {k_value} (rate = {rate_parameter}):", pmf_result)
```

### 11. A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitorswho make a purchase). They collect data from the old and new layouts to compare.
To generate the data use the following command:
import numpy as np
50 purchases out of 1000 visitors
old_layout = np.array([1] * 50 + [0] * 950)
70 purchases out of 1000 visitors  
new_layout = np.array([1] * 70 + [0] * 930)
Apply z-test to find which layout is successful.

```python
import numpy as np
from scipy import stats

# Data from the website layouts
old_layout = np.array([1] * 50 + [0] * 950)  # 50 purchases out of 1000 visitors
new_layout = np.array([1] * 70 + [0] * 930)  # 70 purchases out of 1000 visitors

# Number of successes and trials
n1 = len(old_layout)
n2 = len(new_layout)
x1 = np.sum(old_layout)  # Successes for old layout
x2 = np.sum(new_layout)  # Successes for new layout

# Calculate conversion rates
p1 = x1 / n1  # Conversion rate for old layout
p2 = x2 / n2  # Conversion rate for new layout

# Calculate pooled proportion
p = (x1 + x2) / (n1 + n2)

# Calculate Z statistic
z = (p1 - p2) / np.sqrt(p * (1 - p) * (1/n1 + 1/n2))

# Calculate p-value (one-tailed test)
p_value = stats.norm.cdf(z)

# Results
print(f"Conversion rate for old layout: {p1:.4f}")
print(f"Conversion rate for new layout: {p2:.4f}")
print(f"Z statistic: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Determine success
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis: The new layout has a higher conversion rate.")
else:
    print("Fail to reject the null hypothesis: No significant difference in conversion rates.")
```

```python
import numpy as np
from scipy import stats

# Scores before and after the tutoring program
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

# Calculate means and standard deviations
mean_before = np.mean(before_program)
mean_after = np.mean(after_program)

std_before = np.std(before_program, ddof=1)  # Sample standard deviation
std_after = np.std(after_program, ddof=1)    # Sample standard deviation

n = len(before_program)  # Number of samples

# Calculate Z statistic
z = (mean_after - mean_before) / np.sqrt((std_before ** 2 / n) + (std_after ** 2 / n))

# Calculate the p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

# Results
print(f"Mean before program: {mean_before:.2f}")
print(f"Mean after program: {mean_after:.2f}")
print(f"Z statistic: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Determine if the claim is true or false
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis: The tutoring program improves exam scores.")
else:
    print("Fail to reject the null hypothesis: No significant improvement in exam scores.")
```

### 15. A company is testing two different website layouts to see which one leads to higher click-through rates.
Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of
freedom, and p-value.
Use the following data:
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

```python
import numpy as np
from scipy import stats

def ab_test_analysis(layout_a, layout_b):
    # Calculate sample means
    mean_a = np.mean(layout_a)
    mean_b = np.mean(layout_b)
    
    # Calculate sample variances
    var_a = np.var(layout_a, ddof=1)
    var_b = np.var(layout_b, ddof=1)
    
    # Sample sizes
    n_a = len(layout_a)
    n_b = len(layout_b)
    
    # Calculate t-statistic
    t_statistic = (mean_a - mean_b) / np.sqrt((var_a / n_a) + (var_b / n_b))
    
    # Degrees of freedom
    df = n_a + n_b - 2
    
    # Calculate two-tailed p-value
    p_value = stats.t.sf(np.abs(t_statistic), df) * 2  # Multiply by 2 for a two-tailed test
    
    # Output results
    print(f"Mean of Layout A: {mean_a:.2f}")
    print(f"Mean of Layout B: {mean_b:.2f}")
    print(f"T-statistic: {t_statistic:.4f}")
    print(f"Degrees of freedom: {df}")
    print(f"P-value: {p_value:.4f}")
    
    # Conclusion
    alpha = 0.05  # significance level
    if p_value < alpha:
        print("Reject the null hypothesis: There is a significant difference in click-through rates between Layout A and Layout B.")
    else:
        print("Fail to reject the null hypothesis: No significant difference in click-through rates between Layout A and Layout B.")

# Provided data
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Perform A/B test analysis
ab_test_analysis(layout_a_clicks, layout_b_clicks)
```

### 16. A pharmaceutical company wants to determine if a new drug is more effective than an existing drug in reducing cholesterol levels. Create a program to analyze the clinical trial data and calculate the tstatistic and p-value for the treatment effect.
Use the following data of cholestrol level:
existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]

```python
import numpy as np
from scipy import stats

def drug_effectiveness_analysis(existing_levels, new_levels):
    # Calculate sample means
    mean_existing = np.mean(existing_levels)
    mean_new = np.mean(new_levels)
    
    # Calculate sample variances
    var_existing = np.var(existing_levels, ddof=1)
    var_new = np.var(new_levels, ddof=1)
    
    # Sample sizes
    n_existing = len(existing_levels)
    n_new = len(new_levels)
    
    # Calculate t-statistic
    t_statistic = (mean_existing - mean_new) / np.sqrt((var_existing / n_existing) + (var_new / n_new))
    
    # Degrees of freedom
    df = n_existing + n_new - 2
    
    # One-tailed p-value
    p_value = stats.t.cdf(t_statistic, df)  # Using CDF since it's a one-tailed test
    
    # Output results
    print(f"Mean cholesterol level (existing drug): {mean_existing:.2f}")
    print(f"Mean cholesterol level (new drug): {mean_new:.2f}")
    print(f"T-statistic: {t_statistic:.4f}")
    print(f"Degrees of freedom: {df}")
    print(f"P-value: {p_value:.4f}")
    
    # Conclusion
    alpha = 0.05  # significance level
    if p_value < alpha:
        print("Reject the null hypothesis: The new drug is more effective in reducing cholesterol levels.")
    else:
        print("Fail to reject the null hypothesis: There is no significant evidence that the new drug is more effective.")

# Provided data
existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]

# Perform drug effectiveness analysis
drug_effectiveness_analysis(existing_drug_levels, new_drug_levels)
```

pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

```python
import numpy as np
from scipy import stats

def intervention_analysis(pre_scores, post_scores):
    # Calculate the differences between pre- and post-intervention scores
    differences = np.array(post_scores) - np.array(pre_scores)
    
    # Calculate mean and standard deviation of the differences
    mean_diff = np.mean(differences)
    std_diff = np.std(differences, ddof=1)
    
    # Number of observations
    n = len(differences)
    
    # Calculate t-statistic
    t_statistic = mean_diff / (std_diff / np.sqrt(n))
    
    # One-tailed p-value
    p_value = 1 - stats.t.cdf(t_statistic, df=n-1)
    
    # Output results
    print(f"Mean of differences: {mean_diff:.2f}")
    print(f"T-statistic: {t_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")
    
    # Conclusion
    alpha = 0.05  # significance level
    if p_value < alpha:
        print("Reject the null hypothesis: The intervention had a significant positive impact on math scores.")
    else:
        print("Fail to reject the null hypothesis: There is no significant evidence that the intervention improved scores.")

# Provided data
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

# Perform intervention analysis
intervention_analysis(pre_intervention_scores, post_intervention_scores)
```

### 21. A political analyst wants to determine if there is a significant association between age groups and voter
preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify
them into different age groups and candidate preferences. Perform a Chi-Square test to determine if
there is a significant association between age groups and voter preferences.
Use the below code to generate data:
np.random.seed(0)
age_groups = np.random.choice([ 18 30 , 31 50 , 51+', 51+'], size=30)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Generate the data
np.random.seed(0)
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=30)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

# Create a DataFrame to organize the data
data = pd.DataFrame({'Age Group': age_groups, 'Voter Preference': voter_preferences})

# Create a contingency table
contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

# Perform Chi-Square Test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

# Output results
print("Contingency Table:\n", contingency_table)
print(f"\nChi-Square Statistic: {chi2:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")

# Conclusion based on p-value
alpha = 0.05  # significance level
if p_value < alpha:
    print("\nReject the null hypothesis: There is a significant association between age groups and voter preferences.")
else:
    print("\nFail to reject the null hypothesis: No significant association between age groups and voter preferences.")
```
